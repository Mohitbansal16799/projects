```python
## missing values
dataset.isnull().sum()
```

```python
features_with_na = [features for features in dataset.columns
if dataset[features].isnull().sum()>1]
```

```python
##these are features with nan values
features_with_na
```

```python
#percetnage missing values
for feature in features_with_na:
    print(feature,np.round(dataset[feature].isnull().mean(),4),"% missing values")
```

```python
### lets plot some amazing diagram
data = dataset.copy()
for feature in features_with_na:
    #lets make a variable indicates 1 if there was missing value is observed otherwise 0.
    data[feature]=np.where(data[feature].isnull(),1,0)
    #lets calculatethe mean saleprice where the information is missing or available
    data.groupby(feature)["SalePrice"].median().plot.bar()
    plt.title(feature)
    plt.show()
```

here with relation between the missing values and dependent variable is clearly visible. so we need to replace these nan value with something meaningful which we will do in feature engineering section

```python
datetime_feature=[]
for features in numerical_dataset:
    if 'Yr'in features or "Year" in features:
        datetime_feature.append(features)
print(datetime_feature) 
```

```python
dataset.groupby("YrSold")["SalePrice"].median().plot()
plt.xlabel("yrsold")
plt.ylabel("median of sawlr price")

```

```python
## here we will compare the difference between all year features with sale price
data = dataset.copy()
for features in datetime_feature:
    if features != "YrSold":
        a= data["YrSold"]-data[features]
        plt.scatter(a,data["SalePrice"])
        plt.xlabel(features)
        plt.ylabel("saleprice")
        plt.show()
```

### observation

```python
## numarical variable are two type 
#1.continous variable and 2. discrete variable 
discrete_feature = [feature for feature in numerical_dataset if len(dataset[feature].unique())<=25]
```

```python
len(discrete_feature)
```

```python
dataset[discrete_feature].head()
```

```python
## lets find the realtion between descrete price and sale price 
data =  dataset.copy()
for feature in discrete_feature:
    data.groupby(feature)["SalePrice"].median().plot.bar()
    plt.xlabel("feature")
    plt.ylabel("saleproce")
    plt.show()
    
```

```python
## there is realtion between discrete price and sale price
```

```python
## continous variable
continous_feature = [feature for feature in numerical_dataset if feature not in discrete_feature + datetime_feature ]
```

```python
 len(continous_feature)
```

```python
### lets analyse the continous values by creating histogram to understand the distribution
data = dataset.copy()
for feature in continous_feature:
    data[feature].hist(bins = 25)
    plt.xlabel("feature")
    plt.ylabel("count")
    plt.title(feature)
    plt.show()
```

### EDA part 2


```python
## we will be using logarithmic transformation
data = dataset.copy()
for feature in continous_feature:
    if 0 in data[feature].unique():
        pass
    else :
        data[feature] = np.log(data[feature])
        plt.scatter(data[feature],data["SalePrice"])
        plt.xlabel(feature)
        plt.ylabel("Saleprice")
        plt.title(feature)
        plt.show()
```

```python
##outlier
for feature in continous_feature:
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column = feature)
        plt.ylabel(feature)
        plt.xlabel(feature)
        plt.show()
```
